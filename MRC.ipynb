```python
import matplotlib.pyplot as plt
correlation_matrix = data[list(set(data.columns) - set(['year','artists','id','name','release_date']))].corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.tight_layout()
plt.show()
```

```python
mean=data[list(set(data.columns) - set(['year','artists','id','name','release_date','duration_ms']))].mean()
var= data[list(set(data.columns) - set(['year','artists','id','name','release_date','duration_ms']))].std()
((data[list(set(data.columns) - set(['year','artists','id','name','release_date','duration_ms']))]-mean)/var).plot()
```

```python
# Plotting a box plot of pedestrian counts by weather summary
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[list(set(data.columns) - set(['year','artists','id','name','release_date','duration_ms']))])

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
